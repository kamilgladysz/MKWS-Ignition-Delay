```python
import cantera as ct
import numpy as np
import matplotlib.pyplot as plt
```

```python
gas1 = ct.Solution('gri30.xml')
```

```python
gas1.TPX = 1000, 3*ct.one_atm, 'H2:2, O2:1, N2:3.76'
```

```python
r = ct.Reactor(gas1)
```

```python
sim = ct.ReactorNet([r])
time = 0
times = np.zeros(200)
data = np.zeros([200,4])
```

```python
print ('New test')

print('%10s %10s %10s %14s' % ('t [s]', 'T [K]', 'vol [m3]', 'u [J/kg]'))
for n in range(200):
    time += 1.e-3
    sim.advance(time)
    times[n] = time
    data[n,0] = r.T
    print('%10.3e %10.3f %10.3f %14.6e' % (sim.time, r.T, r.thermo.v, r.thermo.u))
    
plt.plot(times,data[:,0])
plt.xlabel('Time (s)')
plt.ylabel('Temperature (K)')
title = 'Reduced mechanism ignition delay H2, T=1000K, P= three atm'
from textwrap import wrap
plt.title('\n'.join(wrap(title,33)))
plt.rc('xtick', labelsize=1)
plt.show()
```
